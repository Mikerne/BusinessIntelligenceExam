```python
# data_exploration_full.py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from helper_functions import (
    show_boxplots, show_histograms, show_scatter_matrix,
    show_correlation_heatmap, show_grouped_histograms, show_binned_data
)

# ----------------------------
# 1. Load Clean Data
# ----------------------------
df = pd.read_csv("../data/processed/heart_disease_clean_v2.csv")

# ----------------------------
# 2. Define Column Groups
# ----------------------------
continuous_cols = ['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
binary_cols = ['male', 'currentSmoker', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'TenYearCHD']

# ----------------------------
# 3. Summary Tables
# ----------------------------
# Continuous variables
summary_stats = df[continuous_cols].describe().T
summary_stats['IQR'] = summary_stats['75%'] - summary_stats['25%']
print("=== Continuous Variables Summary ===")
print(summary_stats)

# Binary variables
binary_summary = pd.DataFrame({col: df[col].value_counts(normalize=True) * 100 for col in binary_cols}).T
print("\n=== Binary Variables Summary (% of population) ===")
print(binary_summary)

# ----------------------------
# 4. Visualizations
# ----------------------------

# Boxplots
show_boxplots(df[continuous_cols], layout="grid")
print("Boxplots show distribution and outliers for continuous variables.")

# Histograms
show_histograms(df[continuous_cols], bins=15, layout="grid", bell_curve=True)
print("Histograms show distribution and density for continuous variables.")

# Scatter matrices
show_scatter_matrix(df[['age', 'cigsPerDay', 'BMI', 'heartRate']], figsize=(10,10), diagonal='kde')
show_scatter_matrix(df[['sysBP', 'diaBP', 'totChol', 'glucose']], figsize=(10,10), diagonal='kde')
print("Scatter matrices show pairwise relationships between numeric features.")

# Correlation heatmap
show_correlation_heatmap(df[continuous_cols + ['TenYearCHD']])
print("Correlation heatmap shows correlations between variables and CHD.")

# Grouped histograms by CHD
for col in continuous_cols:
    show_grouped_histograms(df[[col, 'TenYearCHD']], category_col='TenYearCHD', bins=15, layout='separate')
    print(f"Grouped histogram for {col} shows differences between CHD risk groups.")

# Violin plots for continuous vars by CHD
for col in continuous_cols:
    plt.figure(figsize=(6,4))
    sns.violinplot(x='TenYearCHD', y=col, data=df, palette='Set2')
    plt.title(f'{col} distribution by 10-Year CHD Risk')
    plt.xlabel('TenYearCHD (0=No, 1=Yes)')
    plt.ylabel(col)
    plt.show()
print("Violin plots show distribution differences between CHD risk groups.")

# Binary variable analysis
for col in binary_cols[:-1]:  # exclude TenYearCHD
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col)
    plt.title(f'Count plot for {col}')
    plt.ylabel('Count')
    plt.show()
print("Count plots show population distribution for binary variables.")

# Proportion of TenYearCHD per binary variable
for binary_col in binary_cols[:-1]:
    cross = pd.crosstab(df[binary_col], df['TenYearCHD'], normalize='index')
    ax = cross.plot(kind='bar', stacked=True, figsize=(6,4), color=['skyblue', 'salmon'])
    for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy()
        ax.text(x + width/2, y + height/2, f'{height*100:.1f}%', ha='center', va='center', fontsize=10, color='black')
    plt.title(f'Proportion of TenYearCHD per {binary_col}')
    plt.ylabel('Proportion')
    plt.xlabel(binary_col)
    plt.legend(title='TenYearCHD', labels=['No', 'Yes'])
    plt.show()
print("Stacked bar plots show CHD risk proportion per binary category.")

# Binned data example
show_binned_data(df, column_to_bin='age', column_to_plot='BMI', bins=5, aggregation_method='mean')
print("Binned data shows BMI changes with age, helping identify high-risk age groups.")
```
