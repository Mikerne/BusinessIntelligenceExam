```python
# report.ipynb
# ----------------------------
# 1. Import Libraries & Data
# ----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys, os

# Tilf√∏j helper functions
sys.path.append(os.path.abspath("../src"))
from helper_functions import (
    show_boxplots, show_histograms, show_scatter_matrix,
    show_correlation_heatmap, show_grouped_histograms, show_binned_data
)

# Load cleaned data
df = pd.read_csv("../data/processed/heart_disease_clean.csv")

# Select relevant columns for the analysis
relevant_cols = [
    'male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 
    'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 
    'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose', 'TenYearCHD'
]
df = df[relevant_cols]

# At this point, data is already cleaned; no further missing value handling is needed.
```

# Report: Predictive Factors for Lifestyle-Related Diseases

**Objective:**  
To identify which lifestyle and patient history factors are most predictive of disease risks, focusing on 10-year risk of CHD.

**Hypothesis H1:**  
A combination of demographic data, lifestyle data, and health records can predict an increased risk of lifestyle-related diseases (e.g., diabetes, heart disease).


## Step 1: Lifestyle & Patient History

We first look at binary variables related to lifestyle and patient history:  
- male  
- currentSmoker  
- BPMeds (blood pressure medication)  
- prevalentStroke  
- prevalentHyp (hypertension)  
- diabetes  

Goal: see their distribution and potential link to 10-year CHD risk.


```python
# Countplots for binary factors
binary_cols = ['male', 'currentSmoker', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes']

for col in binary_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col)
    plt.title(f'Count plot for {col}')
    plt.ylabel("Count")
    plt.grid(axis='y', alpha=0.3)
    plt.show()
```

```python
# Proportion of TenYearCHD per binary factor
for binary_col in binary_cols:
    cross = pd.crosstab(df[binary_col], df['TenYearCHD'], normalize='index')
    cross.plot(kind='bar', stacked=True, figsize=(6,4), color=['skyblue', 'salmon'])
    plt.title(f'Proportion of TenYearCHD per {binary_col}')
    plt.ylabel('Proportion')
    plt.xlabel(binary_col)
    plt.legend(title='TenYearCHD')
    plt.show()
```

**Insight:**  
- Smoking, hypertension, and diabetes show clear differences in 10-year CHD proportion.  
- Male patients also show slightly higher risk.  
- Stroke history and BP meds are less prevalent but may still contribute to risk.


## Step 2: Continuous Features

Examine numeric variables linked to lifestyle and health:  
- Age, cigsPerDay, BMI, heartRate, sysBP, diaBP, totChol, glucose


```python
# Boxplots and Histograms
continuous_cols = ['age', 'cigsPerDay', 'BMI', 'heartRate', 'sysBP', 'diaBP', 'totChol', 'glucose']

show_boxplots(df[continuous_cols], layout="grid")
show_histograms(df[continuous_cols], bins=15, layout="grid", bell_curve=True)
```

## Comments on Boxplots and Histograms

### 1. Age
**Boxplot:** The median age is centered in the box; whiskers show the minimum and maximum without extreme outliers.  
**Histogram:** Age distribution is roughly normal with few older participants on the right tail.  

### 2. cigsPerDay
**Boxplot:** Many 0-values (non-smokers), with some extreme outliers having very high cigarette consumption.  
**Histogram:** Skewed distribution towards low cigarette consumption; few patients smoke many cigarettes daily.  

### 3. BMI
**Boxplot:** Median and IQR show most patients; a few extremely high values are outliers.  
**Histogram:** Skewed distribution toward higher BMI.  

### 4. heartRate
**Boxplot:** Typical resting heart rate around 70 bpm; some high outliers.  
**Histogram:** Approximately normal distribution with slight skew toward higher values.  

### 5. sysBP (systolic blood pressure)
**Boxplot:** Typical range with some high outliers.  
**Histogram:** Skewed distribution toward higher values.  

### 6. diaBP (diastolic blood pressure)
**Boxplot:** Median within normal range; few high outliers.  
**Histogram:** Nearly normal distribution.  

### 7. totChol (total cholesterol)
**Boxplot:** Median and spread show most patients; some have very high values.  
**Histogram:** Slightly skewed distribution toward higher cholesterol levels.  

### 8. glucose
**Boxplot:** Median around normal range; some extremely high values as outliers.  
**Histogram:** Skewed distribution toward higher values.  


```python
# Grouped histograms by TenYearCHD
for col in continuous_cols:
    show_grouped_histograms(df[[col, 'TenYearCHD']], category_col='TenYearCHD', bins=15, layout='separate')
```

**Insight:**  
- Higher BMI, sysBP, diaBP, and glucose levels appear more often in the CHD group.  
- Age is a strong factor: older patients have higher 10-year CHD risk.  
- Cigarettes per day among smokers also correlates with CHD risk.


## Step 3: Correlation Analysis

Compute correlation between numeric features and 10-year CHD.


```python
numeric_cols = continuous_cols + ['TenYearCHD']
show_correlation_heatmap(df[numeric_cols])
```

**Insight:**  
- Positive correlation with CHD: age, sysBP, diaBP, BMI, glucose, cigsPerDay  
- Lifestyle + health factors show moderate to strong correlations, supporting H1.  


## Step 4: Scatter Matrix

Visual inspection of relationships between numeric features.


```python
# Group 1: Demographics & Lifestyle
group1 = ['age', 'cigsPerDay', 'BMI', 'heartRate']
show_scatter_matrix(df[group1], figsize=(10,10), diagonal='kde')

# Group 2: Blood Pressure & Cholesterol
group2 = ['sysBP', 'diaBP', 'totChol', 'glucose']
show_scatter_matrix(df[group2], figsize=(10,10), diagonal='kde')
```

**Insight:**  
- Linear trends are visible for age, BMI, sysBP/diaBP with CHD.  
- Scatter plots reinforce patterns seen in grouped histograms and correlation analysis.


# Conclusion

**Which factors are most predictive of disease risk?**

- **Lifestyle:** current smoking, cigarettes per day, BMI  
- **Health history:** hypertension (prevalentHyp), diabetes, blood pressure medication  
- **Demographics:** age, male sex  

**Support for Hypothesis H1:**  
- The combination of demographic, lifestyle, and health features clearly shows increased CHD risk.
